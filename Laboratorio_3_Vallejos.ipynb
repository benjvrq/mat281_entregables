El objetivo es tratar de obtener la mayor información posible de este conjunto de datos. Para cumplir este objetivo debe resolver las siguientes problemáticas:

6.- Describir el conjunto de datos (**hint**: .describe())

**Nota**: Dado el tipo de dato que tiene cada columna, no es posible obtener medidas de tendencia central para el campo "gender" y el campo "occupation".

7.- Imprimir solo la columna de **occupation**.

```python
# Imprimir columna "ocupación"
occupation_col = df['occupation'].reset_index()
occupation_col = occupation_col.drop(['index'], axis=1)
print("\nLa columna 'occupation' es:\n")
occupation_col
```

8.- ¿Cuántas ocupaciones diferentes hay en este conjunto de datos?

```python
# Número de ocupaciones
num_occupation = df['occupation'].nunique()
print("\nEl número de ocupaciones diferentes es:", num_occupation)
```

9.- ¿Cuál es la ocupación más frecuente?

```python
# Ocupación más frecuente
occupation_count = df['occupation'].value_counts()
freq_occupations = occupation_count.loc[occupation_count == occupation_count.max()]
if len(freq_occupations) > 1:
    print("\nLas ocupaciones más frecuentes con un total de:", occupation_count[0], "persona(s) por ocupación, son:\n", freq_occupations)
else:
    print("\nLa ocupación más frecuente es:", occupation_count.index[0],", con un total de:", occupation_count[0])
```

10.- ¿Cuál es la edad media de los usuarios?

```python
# Edad media de los usuarios
age_mean = df['age'].mean()
print("\nLa edad media de los usuarios es:", int(age_mean))
```

**Nota**: En este caso, se ha decidido optar por aproximar al entero más cercano dada la naturaleza de los datos, pues el campo 'age' es del tipo entero.

11.- ¿Cuál es la edad con menos ocurrencia?

```python
# Edad con menos ocurrencia
age_counts = df['age'].value_counts()
freq_age = age_counts.loc[age_counts == age_counts.min()]
if len(freq_age) > 1:
    print("\nLas edades menos frecuentes, con un total de", age_counts.min(), "persona(s) por edad, son:\n", freq_age)
else:
    print("\nLa edad menos frecuente es:", freq_age,", con un total de:", freq_age)
```
