```python
# Importación de la data district_info.csv
district_df = pd.read_csv(path + "districts_info.csv")
district_df.columns = [x.lower().replace(' ','_') for x in district_df.columns]
products_df.head(5)
```

```python
# Importación de engagement_data
path = '/content/drive/MyDrive/MAT281/Tareas/Tarea 2/engagement_data'
all_files = glob.glob(path + "/*.csv")
li = []
all_files

for filename in all_files:
    df = pd.read_csv(filename, index_col=None, header=0,engine=None)
    district_id = filename.split("/")[-1].split(".")[0]
    df["district_id"] = district_id
    li.append(df)

engagement_df = pd.concat(li)
engagement_df = engagement_df.reset_index(drop=True)
engagement_df
```

### Análisis de datos: distritos

Veamos cuantos datos tenemos por distrito, mediante **gráfico de torta:**

```python
# Gráfico de torta: Distribución poblacional
fig, ax  = plt.subplots(figsize=(16, 8))
fig.suptitle('Distrubución de Estados', size = 20)
district_df["state"].value_counts().head(15).plot(kind = 'pie', autopct='%1.1f%%', figsize=(16, 8), startangle=0).legend();
print("\n---------------- Gráfico 1 ----------------\n")
```

```python
# Gráfico de torta: Distribución localidad
print("\n---------------- Gráfico 2 ----------------\n")
fig, ax  = plt.subplots(figsize=(16, 8))
fig.suptitle('Distribución de Localidades', size = 20)
district_df["locale"].value_counts().head(15).plot(kind = 'pie', autopct='%1.1f%%', figsize=(16, 8), startangle=0).legend();
plt.show()
```

Del **gráfico 1** se desprende que gran parte de la población en estudio se concentrar en Massachusetts, Utah y Connecticut.

Del **gráfico 2** observamos que un 18.8% vive en zona rural. Gran parte de la población de estudio no debería tener problemas de conexión a internet, salvo este porcentaje de gente, quienes pueden tener dificultades a la hora de asistir a clases.

```python
# Gráfico de barras: Minorias
print("----------- Gráfico 1 ----------")

fig, ax  = plt.subplots(figsize=(16, 8))
sns.countplot(data= district_df, x = "pct_black/hispanic", label="Black/hispanic");
plt.title("Black/hispanic")
plt.show();
```

```python
# Gráfico de barras: Becas
print("----------- Gráfico 2 ----------")

fig, ax  = plt.subplots(figsize=(16, 8))
sns.countplot(data= district_df, x = "pct_free/reduced");
plt.title("Free/Reduced")
plt.show();
```

Del gráfico 1, notamos entonces que los distritos que tienen hasta un 20% de personas afrodescendientes/latinas, están en mayor cantidad.

Del gráfico 2, notamos que aquellos que no superan un porcentaje de 60% de estudiantes afrodescendientes/latinos, poseen programas de apoyo en la alimentación, como reducciones de precio o bien almuerzos gratis.

### Análisis de datos: distribuidores

Analicemos los datos de las compañías que ofrecen su serivicio de educación online.

```python
# Gráfico de barras: Compañías más solicitadas
plt.figure(figsize = (20,10))
sns.countplot(data=products_df, x="provider/company_name", order=products_df['provider/company_name'].value_counts().index[:10])
plt.title('Top 10 Provider Platform with the Most Product in 2020', size=12)
plt.show()
```

Notamos que Google LLC provee más serivicios, respecto a los otros.

```python
# Gráfico de barras: servicios utilizados
plt.figure(figsize = (16,8))
sns.countplot(x=products_df["primary_essential_function"]);
plt.xticks(rotation=90);
```

Las plataformas de aprendizaje digital superar en creces a los demás serivicios. Le sigue los sitios, recurosos y referencias. Después viene todo el resto de serivicos.

```python
plt.figure(figsize=(16, 10))
B = sns.countplot(y='sector(s)',  #eje y
              data=products_df, #info gráfico
              order=products_df["sector(s)"].value_counts().index[:10], palette="husl") #info gráfico
B.set(xlabel="Suma", ylabel = "sector(s)") #etiquetas ejes
plt.title("Cantidad por sectores")
plt.show()
```

Prekinder y la educación superior son aquellos sectores de la educación que más ocupan serivicios de educación.
